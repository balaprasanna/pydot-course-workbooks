```python
foldername = r"C:\Users\balap\Desktop\teaching\pydot27\Workshops\Workshop03\pyvideo_data-master\pyvideo_data-master"

import json
from pymongo import MongoClient
from pathlib import Path

path = Path(foldername)
files = list(path.glob("**/videos/*.json"))


def get_collection_ref():
    client = MongoClient("localhost:27017")
    collection = client["videosdb"]["videos"]
    return collection

def save_doc_to_mongodb(doc):
    collection.insert_one(doc)

collection = get_collection_ref()

for file in files:
    try:
        #open the file
        with open(file, mode="r", encoding="utf-8") as f:
            doc = json.loads(f.read())
            save_doc_to_mongodb(doc)
    except Exception as e:
        print(e, file)
```
